```python
#Q4 Polynomial plotting with Annotation 
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(39)

# Generate random coefficients for a fifth degree polynomial
coefficients = np.random.rand(6) * 10 - 5  # Random coefficients between -5 and 5

# Define the polynomial function
def polynomial(x, coeffs):
    return np.polyval(coeffs, x)

# Generate x values for plotting
x_values = np.linspace(-5, 5, 400)

# Calculate y values using the polynomial function
y_values = polynomial(x_values, coefficients)

# Find local maxima and minima
maxima_indices = np.where(np.r_[True, y_values[1:] > y_values[:-1]] & np.r_[y_values[:-1] > y_values[1:], True])
minima_indices = np.where(np.r_[True, y_values[1:] < y_values[:-1]] & np.r_[y_values[:-1] < y_values[1:], True])

maxima = (x_values[maxima_indices], y_values[maxima_indices])
minima = (x_values[minima_indices], y_values[minima_indices])

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')

# Annotate maxima
for x, y in zip(*maxima):
    plt.annotate(f'Max\n({x:.2f}, {y:.2f})', xy=(x, y), xytext=(-40, 20), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

# Annotate minima
for x, y in zip(*minima):
    plt.annotate(f'Min\n({x:.2f}, {y:.2f})', xy=(x, y), xytext=(-40, -20), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Fifth Degree Polynomial')
plt.grid(True)
plt.legend()
plt.show()
```
